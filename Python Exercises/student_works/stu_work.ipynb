```python
def optimized_bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr


import time

# Function to measure the running time
def measure_time(arr):
    start_time = time.time()
    optimized_bubble_sort(arr)
    end_time = time.time()
    return end_time - start_time

import matplotlib.pyplot as plt

sizes = [10, 50, 100, 500, 1000]  # Array sizes to test
times_sorted = []
times_reverse = []

for size in sizes:
    sorted_array = list(range(size))
    reverse_array = list(range(size, 0, -1))
    
    time_sorted = measure_time(sorted_array)
    time_reverse = measure_time(reverse_array)
    
    times_sorted.append(time_sorted)
    times_reverse.append(time_reverse)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(sizes, times_sorted, label='Sorted Array')
plt.plot(sizes, times_reverse, label='Reverse Array')
plt.xlabel('Number of Elements')
plt.ylabel('Time (seconds)')
plt.title('Running Time of Optimized Bubble Sort')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import sys
import numpy as np

sys.setrecursionlimit(10**4)

def selectionSort(array):
    size = len(array)
    for ind in range(size):
        min_index = ind

        for j in range(ind + 1, size):
            if array[j] < array[min_index]:
                min_index = j
        (array[ind], array[min_index]) = (array[min_index], array[ind])
    return array

def sortedplot(function: "function", ten: int, col: str, reverse = 1) -> list:
    l = list()
    for i in range(0, ten):
        arr = list(range(0, 10**(i)))
        if not reverse: np.random.shuffle(arr)
        else: arr = arr[::reverse]
        start = time.perf_counter()
        function(arr)
        end = time.perf_counter()
        l.append(end-start)
    plt.plot(l, color = col)
    return l    

if __name__ == "__main__":
    sortedplot(selectionSort, 6, "b", 1)
    sortedplot(selectionSort, 6, "r", -1)
    plt.title("selection")
    plt.legend(["sorte", "reverse"], loc = "upper left")
    plt.xlabel("array size, log base 10")
    plt.ylabel("time in seconds")
    plt.show()
```
