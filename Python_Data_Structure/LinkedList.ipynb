```python
class Node:
    def __init__(self, data):
        self.data = data  
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def insertAtBeginning(self, new_data):
        new_node = Node(new_data)  
        new_node.next = self.head  
        self.head = new_node
    
    def insertAtEnd(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node  
            return
        last = self.head 
        while last.next:  
            last = last.next
        last.next = new_node

    def deleteFromBeginning(self):
        if self.head is None:
            return "The list is empty" 
        self.head = self.head.next

    def deleteFromEnd(self):
        count = 1
        if self.head is None:
            return "The list is empty" 
        if self.head.next is None:
            self.head = None  # If there's only one node, remove the head by making it None
            return
        temp = self.head
        while temp.next.next:  # Otherwise, go to the second-last node
            count +=1            
            temp = temp.next
        temp.next = None
        print(count)

    def search(self, value):
        current = self.head 
        position = 0  
        
        while current:            
            if current.data == value:                 
                return f"Value '{value}' found at position {position}" 
            current = current.next
            position += 1
        return f"Value '{value}' not found in the list after {position} comparisons" 
    
    def reverse(self):        
        prev = None
        current = self.head
        while current:
            next_node = current.next  
            current.next = prev       
            prev = current            
            current = next_node       
        self.head = prev         

    def jump(self, current):
        
        if current is None:
            return "The end" 
        print(current.data)

        if current.next is not None:
            self.jump(current.next.next)
        print(current.data)
       

    def printList(self):
        temp = self.head 
        while temp:
            print(temp.data,end=' ') 
            temp = temp.next 
        print() 
```

```python
## create a linkedlist  'red'->'blue'->'yellow'->'orange'->'green'->'white'

llist = LinkedList()
llist.insertAtEnd('red')
llist.insertAtEnd('blue')
llist.insertAtEnd('yellow')
llist.insertAtEnd('orange')
llist.insertAtEnd('green')
llist.insertAtEnd('white')
llist.printList()
```
