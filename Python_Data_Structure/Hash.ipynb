```python
## Part 1: Hash with collisions
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash_function(self, key):        
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)        
        self.table[index].append((key, value))

    def print_table(self):
        print("Hash Table:")
        print("Index\tKey\tValue")
        for i, slot in enumerate(self.table):
            print(f"{i}\t", end="")
            if slot:
                for key, value in slot:
                    print(f"{key}\t\t{value}")
            print()


hash_table = HashTable(10)
hash_table.insert("John Doe", 30)
hash_table.insert("Jane Smith", 25)
hash_table.insert("Alice Johnson", 35)
hash_table.insert("Bob Brown", 40)
hash_table.insert("Emily Davis", 28)
hash_table.insert("Michael Wilson", 32)

hash_table.print_table()
```
